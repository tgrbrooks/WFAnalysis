```python
import ROOT
import sys, os
import numpy as np
import getChInfo
from root_numpy import root2array, root2rec, tree2rec, list_branches
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline

chinfo = getChInfo.ChanInfo()

f = ROOT.TFile("noise_correlation_run138.root")

ch = f.Get("ch_tree")
corr = f.Get("corr_tree")
ch_arr = tree2rec(ch,branches=['larch','crate','slot','femch','rms','evt','mean','block','adc_v'])
#corr_arr = tree2rec(corr,branches=['ref_larch','sub_larch','rms_ref', 'rms_sub', 'block', 'mean_ref', 'mean_sub', 'mean_diff', 'corr'])
#corr_arr.dtype.names = ('ref_larch','sub_larch','rms_ref', 'rms_sub', 'block', 'mean_ref', 'mean_sub', 'mean_diff', 'corr')

adc_dict = {}
for y in xrange(len(ch_arr)):
    adc_dict[(ch_arr['larch'][y],ch_arr['evt'][y])] = ch_arr['adc_v'][y]-ch_arr['mean'][y]
    
print len(adc_dict)

def plotwf(larch,evt,start,ticks,col='r',lab='WF',lw=1):
    wf = adc_dict[larch,evt]
    p = plt.plot(np.linspace(start,start+ticks,ticks), wf[start:start+ticks],color=col,label=lab,linewidth=lw)
    
def plotdiffwf(larch1,larch2,evt,start,ticks,col='r',lab='WF',lw=1):
    wf1 = adc_dict[larch1,evt]
    wf2 = adc_dict[larch2,evt]
    wfdiff = []
    for x in xrange(ticks):
        wfdiff.append(wf1[start+x]-wf2[start+x])
    p = plt.plot(np.linspace(start,start+ticks,ticks), wfdiff,color=col,label=lab,linewidth=lw)
    
    
def flip(tick1,tick2,tick3):
    if ( (tick3>0) and (tick2 <0) and (tick1 >0) ):
        return -1
    elif ( (tick3<0) and (tick2 >0) and (tick1 <0) ):
        return -1
    return 1

def findZigZagROI(larch,evt,start,ticks):
    wf = adc_dict[larch,evt]
    zigzag = 0
    for x in xrange(ticks-2):
        zigzag += flip(wf[start+x],wf[start+x+1],wf[start+x+2])
    zigzag /= float(ticks-2)
    return zigzag

def findZigZag(larch,evt,minlength=5):
    wf = adc_dict[larch,evt]
    segments = []
    start = 0
    end   = 0
    ticksOn = 0
    for x in xrange(len(wf)-2):
        if (  ( ((wf[x]>0) and (wf[x+1]<0)  and (wf[x+2]>0)) or ((wf[x]<0) and (wf[x+1]>0)  and (wf[x+2]<0)) ) 
            and ( (np.abs(wf[x+1]-wf[x]) >= 3) and (np.abs(wf[x+2]-wf[x+1]) >= 3) ) ):
            if (ticksOn == 0):
                # restart
                start = x
            ticksOn += 1
        else:
            if (ticksOn != 0):
                # close segment
                end = x
                if (ticksOn >= minlength):
                    segments.append([start,end])
            ticksOn = 0
    return segments

def rms(l,pl):
    if (pl==1):
        return (14/4.7)*(1.16+0.0337*l+0.00639*l*l)
    return (14/4.7)*(1.19+0.0442*l+0.00219*l*l)
```
