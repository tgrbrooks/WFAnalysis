```python
import ROOT
import sys, os
import numpy as np
import getChInfo
from root_numpy import root2array, root2rec, tree2rec, list_branches
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline

plt.rcParams.update({'font.size': 16})

chinfo = getChInfo.ChanInfo()

f = ROOT.TFile("../noise_zigzag_run002_ev6_CorrMatrix.root")

ch = f.Get("ch_tree")
corr = f.Get("corr_tree")
corr_arr = tree2rec(corr,branches=['ref_larch','sub_larch','rms_ref', 'rms_sub', 'block', 'mean_ref', 'mean_sub', 'mean_diff', 'corr','corrall'],selection='block==0')
ch_arr = tree2rec(ch,branches=['larch','crate','slot','femch','rms','evt','mean','block','adc_v','zigzag_start','zigzag_end'])
```

```python
chan_list = [1103,4175,1102,4174,1101,4173,1100,4172,1099,4171,1098,4170,1097,4169,1096,4168,1711,4783,1710,4782,1709,4781,1708,4780,1707,4779,1706,4778,1705,4777,1704,4776,8255,8254,8253,8252,8251,8250,8249,8248,8247,8246,8245,8244,8243,8242,8241,8240]
ch_list_1 = [1103,4175,1102,4174,1101,4173,1100,4172,1099,4171,1098,4170,1097,4169,1096,4168]
ch_list_2 = [1711,4783,1710,4782,1709,4781,1708,4780,1707,4779,1706,4778,1705,4777,1704,4776]
ch_list_3 = [8255,8254,8253,8252,8251,8250,8249,8248,8247,8246,8245,8244,8243,8242,8241,8240]
print len(corr_arr['corr'])
#plt.hist(corr_arr['corr'])
#plt.show()
```

```python
#chan_list = ch_list_3

mat = np.zeros([len(chan_list),len(chan_list)])
mat2 = np.zeros([len(chan_list),len(chan_list)])

mat_l1 = np.zeros([len(ch_list_1),len(ch_list_1)])
mat2_l1 = np.zeros([len(ch_list_1),len(ch_list_1)])

mat_l2 = np.zeros([len(ch_list_2),len(ch_list_2)])
mat2_l2 = np.zeros([len(ch_list_2),len(ch_list_2)])

mat_l3 = np.zeros([len(ch_list_3),len(ch_list_3)])
mat2_l3 = np.zeros([len(ch_list_3),len(ch_list_3)])

for i in xrange(len(corr_arr['ref_larch'])):
    refch = int(corr_arr['ref_larch'][i])
    subch = int(corr_arr['sub_larch'][i])
    # if in general list
    if (  ((refch in chan_list) == True) and ((subch in chan_list) == True) ):
        refidx = chan_list.index(refch)
        subidx = chan_list.index(subch)
        if (refidx == subidx):
            mat[refidx][subidx] = np.nan
            mat2[refidx][subidx] = np.nan
        else:
            mat[refidx][subidx] = float(corr_arr['corr'][i])
            mat2[refidx][subidx] = float(corr_arr['corrall'][i])
    # if in list for ASIC 1
    if (  ((refch in ch_list_1) == True) and ((subch in ch_list_1) == True) ):
        refidx = ch_list_1.index(refch)
        subidx = ch_list_1.index(subch)
        if (refidx == subidx):
            mat_l1[refidx][subidx] = np.nan
            mat2_l1[refidx][subidx] = np.nan
        else:
            mat_l1[refidx][subidx] = float(corr_arr['corr'][i])
            mat2_l1[refidx][subidx] = float(corr_arr['corrall'][i])
    # if in list for ASIC 2
    if (  ((refch in ch_list_2) == True) and ((subch in ch_list_2) == True) ):
        refidx = ch_list_2.index(refch)
        subidx = ch_list_2.index(subch)
        if (refidx == subidx):
            mat_l2[refidx][subidx] = np.nan
            mat2_l2[refidx][subidx] = np.nan
        else:
            mat_l2[refidx][subidx] = float(corr_arr['corr'][i])
            mat2_l2[refidx][subidx] = float(corr_arr['corrall'][i])
    # if in list for ASIC 3
    if (  ((refch in ch_list_3) == True) and ((subch in ch_list_3) == True) ):
        refidx = ch_list_3.index(refch)
        subidx = ch_list_3.index(subch)
        if (refidx == subidx):
            mat_l3[refidx][subidx] = np.nan
            mat2_l3[refidx][subidx] = np.nan
        else:
            mat_l3[refidx][subidx] = float(corr_arr['corr'][i])
            mat2_l3[refidx][subidx] = float(corr_arr['corrall'][i])


```

```python
fig, ax = plt.subplots(figsize=(6,6))
matpic = ax.matshow(mat_l3)
pos = fig.add_axes([0.95,0.15,0.05,0.7])
cbar = plt.colorbar(matpic,cax=pos)
cbar.set_clim(vmin=-0.5,vmax=0.5)
fig.show()

fig, ax = plt.subplots(figsize=(6,6))
matpic = ax.matshow(mat2_l3)
pos = fig.add_axes([0.95,0.15,0.05,0.7])
cbar = plt.colorbar(matpic,cax=pos)
cbar.set_clim(vmin=-0.3,vmax=0.3)
plt.show()
```
