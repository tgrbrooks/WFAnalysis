```python
# find number of zig-zag channels affected vs. time
nevts = 30
nchans_zigzag = np.ndarray((nevts,9600),dtype=int)

for n in xrange(nevts):
    for m in xrange(9600):
        nchans_zigzag[n][m] = 0

for x in xrange(len(ch_arr['larch'])): 
    starts = ch_arr['zigzag_start'][x]
    ends   = ch_arr['zigzag_end'][x]
    evt    = ch_arr['evt'][x]
    for n in xrange(len(starts)):
            for j in xrange(ends[n]-starts[n]):
                nchans_zigzag[evt][starts[n]+j] += 1
```

```python
# returns a list of:
# [start, end, [ [maxTick,maxAmp], [maxTick,maxAmp], [maxTick,maxAmp], ...]]
# each object defines a pulse wtih possibly several packets inside

def findROIs(vec,thresh):
    maxmax = 0
    found = False
    s = 0
    e = 0
    maxtick = 0
    maximum = 0
    pulses = []
    maxima = []
    # rising: if we have encountered a region in which the packet is rising
    rising = False
    thesemax = []
    for i in xrange(len(vec)):
        if (vec[i] > thresh):
            if (vec[i] > maxmax):
                maxmax = vec[i]
            if (found == False):
                s = i
                # reset maxima for this pulse
                thesemax = []
            found = True
            if (vec[i] > vec[i-1]):
                maximum = vec[i]
                maxtick = i
            else:
                # see if we have fallen down enough to claim we have passed a peak
                if ( (maximum-vec[i]) > 20):
                    thesemax.append([maxtick,maximum])
                    maximum = 0
        if (vec[i] < thresh):
            if (found == True):
                e = i
                if (maxmax > 200):
                    pulses.append([s,e,thesemax])
            maxmax = 0
            found = False
            thesemax = []
    return pulses
```

```python
#interesting threshold values: 200, 500, 10

widths = []
spacings = []
num_maxima_in_pulse = []
spacing_maxima_in_pulse = []
sub_burst_peak = []
for x in xrange(len(nchans_zigzag)):
    pulses = findROIs(nchans_zigzag[x],10)
    for n in xrange(1,len(pulses)):
        widths.append(pulses[n][1]-pulses[n][0])
        if (n != 0):
            spacings.append(pulses[n][0]-pulses[n-1][1])
            num_maxima_in_pulse.append(len(pulses[n][2]))
            for j in xrange(len(pulses[n][2])-1):
                spacing_maxima_in_pulse.append(pulses[n][2][j+1][0]-pulses[n][2][j][0])
                sub_burst_peak.append(pulses[n][2][j+1][1])
            
            
w = np.array(widths)
s = np.array(spacings)
plt.xlabel('Width of each Burst [Ticks]',fontsize=18)
plt.hist(w,bins=40)
plt.show()
plt.xlabel('Spacing Between Bursts [Ticks]',fontsize=18)
plt.hist(s,bins=40)
plt.show()
plt.xlabel('Number of Maxima in each Burst',fontsize=18)
plt.hist(num_maxima_in_pulse)
plt.show()
plt.xlabel('Spacing Between Maxima in each Burst [Ticks]',fontsize=18)
plt.hist(spacing_maxima_in_pulse,bins=20)
#plt.xticks().set_fontsize(18)
plt.show()
plt.xlabel('Maximum number of Channels Affected in each Sub-Burst',fontsize=18)
plt.hist(sub_burst_peak,bins=30)
#plt.xticks().set_fontsize(18)
plt.show()
```
