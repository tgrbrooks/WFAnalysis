```python
import ROOT
import sys, os
import numpy as np
import getChInfo
from root_numpy import root2array, root2rec, tree2rec, list_branches
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline

plt.rcParams.update({'font.size': 16})

chinfo = getChInfo.ChanInfo()

run = 151
sstart = 0
send = 9
gain = 14.

fname = "../highstat_5tick/noise_zigzag_run%03i_%03i_%03i.root"%(run,sstart,send)
print fname
f = ROOT.TFile(fname)

zig = f.Get("zig_tree")
zig_arr = tree2rec(zig,branches=['evt','bursts','frac'])
zz = f.Get("z_tree")
zz_arr = tree2rec(zz,branches=['evt','start','end','amp','larch','crate','slot','femch','len'])
```

```python
# returns a list of:
# [start, end, [ [maxTick,maxAmp], [maxTick,maxAmp], [maxTick,maxAmp], ...]]
# each object defines a pulse wtih possibly several packets inside

def findBursts(vec,thresh,minmax):
    maxmax = 0
    found = False
    s = 0
    e = 0
    maxtick = 0
    maximum = 0
    pulses = []
    maxima = []
    # rising: if we have encountered a region in which the packet is rising
    rising = False
    thesemax = []
    for i in xrange(len(vec)):
        if (vec[i] > thresh):
            if (vec[i] > maxmax):
                maxmax = vec[i]
            if (found == False):
                s = i
                # reset maxima for this pulse
                thesemax = []
            found = True
            if (vec[i] > vec[i-1]):
                maximum = vec[i]
                maxtick = i
            else:
                # see if we have fallen down enough to claim we have passed a peak
                if ( (maximum-vec[i]) > 5):
                    thesemax.append([maxtick,maximum])
                    maximum = 0
        if (vec[i] < thresh):
            if (found == True):
                e = i
                if (maxmax > minmax):
                    pulses.append([s,e,thesemax])
            maxmax = 0
            found = False
            thesemax = []
    return pulses
```

```python
widths = []
spacings = []
num_maxima_in_pulse = []
spacing_maxima_in_pulse = []
sub_burst_peak = []
for x in xrange(len(zig_arr['bursts'])):
    pulses = findBursts(zig_arr['bursts'][x],2,100)
    for n in xrange(1,len(pulses)):
        widths.append(pulses[n][1]-pulses[n][0])
        if (n != 0):
            spacings.append(pulses[n][0]-pulses[n-1][1])
            num_maxima_in_pulse.append(len(pulses[n][2]))
            for j in xrange(len(pulses[n][2])-1):
                spacing_maxima_in_pulse.append(pulses[n][2][j+1][0]-pulses[n][2][j][0])
                sub_burst_peak.append(pulses[n][2][j+1][1])    
```

```python
fig,ax = plt.subplots(figsize=(8,6))
w = np.array(widths)
s = np.array(spacings)
plt.xlabel('Width of each Burst [Ticks]',fontsize=18)
plt.xlim([0,100])
plt.hist(w,bins=np.linspace(0,100,100),label='Run %03i'%run)
#plt.show()
plt.legend(loc=1)
plt.savefig("plots/burst_width_%03i_%03i_%03i.pdf"%(run,sstart,send))

fig,ax = plt.subplots(figsize=(8,6))
plt.xlabel('Spacing Between Bursts [Ticks]',fontsize=18)
plt.xlim([0,500])
plt.hist(s,bins=np.linspace(0,500,50),label='Run %03i'%run)#,log='True')
plt.legend(loc=1)
#plt.show()
plt.savefig("plots/burst_spacing_%03i_%03i_%03i.pdf"%(run,sstart,send))

fig,ax = plt.subplots(figsize=(8,6))
plt.xlabel('Number of Maxima in each Burst',fontsize=18)
plt.hist(num_maxima_in_pulse,bins=np.linspace(0.5,10.5,11),label='Run %03i'%run)
plt.xlim([0,10])
#plt.show()
plt.legend(loc=1)
plt.savefig("plots/num_maxima_%03i_%03i_%03i.pdf"%(run,sstart,send))

fig,ax = plt.subplots(figsize=(8,6))
plt.xlabel('Spacing Between Maxima in each Burst [Ticks]',fontsize=18)
plt.hist(spacing_maxima_in_pulse,bins=np.linspace(0.5,30.5,31),label='Run %03i'%run)
plt.xlim([0,30])
#plt.show()
plt.legend(loc=1)
plt.savefig("plots/maxima_spacing_%03i_%03i_%03i.pdf"%(run,sstart,send))

#plt.hist(sub_burst_peak,bins=np.linspace(0,3500,100),log='True',label='Run %03i'%run)
#plt.legend(loc=1)
#plt.show()
```
