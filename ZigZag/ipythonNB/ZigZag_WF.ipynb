```python
import ROOT
import sys, os
import numpy as np
import getChInfo
from root_numpy import root2array, root2rec, tree2rec, list_branches
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline

chinfo = getChInfo.ChanInfo()

plt.rcParams.update({'font.size': 16})

run = 95
sstart = 0
send = 0
gain = 14

fname = "../zigzagwfs/noise_zigzag_run%03i_%03i_%03i.root"%(run,sstart,send)
print fname
f = ROOT.TFile(fname)

ch = f.Get("ch_tree")
ch_arr = tree2rec(ch,branches=['larch','crate','slot','femch','rms','evt','mean','block','adc_v','zigzag_start','zigzag_end'])
```

```python
crates = [[],[],[]]
wires  = [[],[],[]]
for x in xrange(len(ch_arr['larch'])):
    crate = ch_arr['crate'][x]
    slot  = ch_arr['slot'][x]
    femch = ch_arr['femch'][x]
    l = chinfo.getlarch(crate,slot,femch)
    if (l is not None):
        if (l.plane == 0):
            crates[0].append(crate)
            wires[0].append(l.wirenum)
        if (l.plane == 1):
            crates[1].append(crate)
            wires[1].append(l.wirenum)
        if (l.plane == 2):
            crates[2].append(crate)
            wires[2].append(l.wirenum)
        
```

```python
fig=plt.figure(figsize=(8,8))
plt.plot(crates[2],wires[2],'bo')
plt.show()
```

```python
adc_dict = {}
for y in xrange(len(ch_arr)):
    adc_dict[(ch_arr['larch'][y],ch_arr['evt'][y])] = ch_arr['adc_v'][y]-ch_arr['mean'][y]
```

```python
def plotwf(chan,ev,start,end,c='b',lw=1,l=''):
    if (l==''):
        plt.plot(np.linspace(start,end,end-start),adc_dict[(chan,ev)][start:end],color=c,linewidth=lw)
    else:
        plt.plot(np.linspace(start,end,end-start),adc_dict[(chan,ev)][start:end],color=c,linewidth=lw,label=l)
```
