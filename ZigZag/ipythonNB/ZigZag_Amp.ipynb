```python
# first find which channels have an amplitude larger than ampCut at least once

pos_dict = {}

ampCut = 70

for x in xrange(len(zz_arr['larch'])):
    amp = (14./gain)*zz_arr['amp'][x]
    if (amp > ampCut):
        crate = zz_arr['crate'][x]
        slot = zz_arr['slot'][x]
        femch = zz_arr['femch'][x]
        CSF = [crate,slot,femch]
        larch = zz_arr['larch'][x]
        pos_dict[larch] = CSF
```

```python
#  now keep track of all amplitudes for those channels

amp_dict = {}
for x in xrange(len(zz_arr['larch'])):
    larch = zz_arr['larch'][x]
    if ( (larch in pos_dict) == True):
        amp = zz_arr['amp'][x]
        if ( (larch in amp_dict) == False):
            amp_dict[larch] = [amp]
        else:
            amp_dict[larch].append(amp)
```

```python
fout = open('run%03i_%03i_%03i_highAmp_zigzag.txt'%(run,sstart,send),'w')
outstr = "Chan\tCrate\t\Slot\Femch\tCut\tNpass\n"
fout.write(outstr)
for n in amp_dict:
    csf = pos_dict[n]
    if (len(amp_dict[n]) > 100):
        outstr =  "%i\t%i\t%i\t%i\t%i\t%i\n"%(n,csf[0],csf[1],csf[2],ampCut,len(amp_dict[n]))
        print outstr
        fout.write(outstr)
fout.close()
```
