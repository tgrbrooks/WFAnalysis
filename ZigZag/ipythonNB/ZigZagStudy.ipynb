```python
import ROOT
import sys, os
import numpy as np
import getChInfo
from root_numpy import root2array, root2rec, tree2rec, list_branches
import matplotlib.pyplot as plt
from matplotlib import colors
%matplotlib inline

chinfo = getChInfo.ChanInfo()

plt.rcParams.update({'font.size': 16})

run = 2
sstart = 0
send = 0
gain = 4.7

fname = "../zigzagwfs/noise_zigzag_run%03i_%03i_%03i.root"%(run,sstart,send)
print fname
f = ROOT.TFile(fname)

ch = f.Get("ch_tree")
ch_arr = tree2rec(ch,branches=['larch','crate','slot','femch','rms','evt','mean','block','adc_v','zigzag_start','zigzag_end'])
```

```python
adc_dict = {}
for y in xrange(len(ch_arr)):
    adc_dict[(ch_arr['larch'][y],ch_arr['evt'][y])] = ch_arr['adc_v'][y]-ch_arr['mean'][y]
```

```python
#ig,ax = plt.subplots(figsize=(10,6))
tick = []
adc = []
adc_max = []
matadcmax = 70
mattickmax = 30
mat = np.zeros([matadcmax,mattickmax])
zzLength = []
for x in xrange(len(ch_arr['larch'])): 
    if (ch_arr['larch'][x] >= 0):
        chan = ch_arr['larch'][x]
        starts = ch_arr['zigzag_start'][x]
        ends   = ch_arr['zigzag_end'][x]
        for s in xrange(len(starts)):
            zzLength.append(ends[s]-starts[s])
            thismax = 0
            adc_vec = []
            tick_vec = []
            for n in xrange(ends[s]-starts[s]):
                adc = adc_dict[chan,0][starts[s]+n]
                abs_adc = int(np.abs((14./gain)*adc))
                #adc_vec.append(adc)
                #tick_vec.append(n)
                if (abs_adc > thismax):
                    thismax = abs_adc
                if ((abs_adc < matadcmax) and (n < mattickmax)):
                    mat[abs_adc][n] += 1
            #plt.plot(tick_vec,adc_vec)
            #plt.show()
            adc_max.append(thismax)
```

```python
from matplotlib.colors import LogNorm

fig, ax = plt.subplots(figsize=(10,8))
#pos = fig.add_axes([0.95,0.15,0.05,0.7])
matpic = ax.matshow(mat,origin='lower',aspect='auto',norm=LogNorm(vmin=1,vmax=50000))
plt.xlabel('Time-Tick',fontsize=18)
plt.ylabel('abs [ ADC - Baseline ]',fontsize=18)
plt.title('Zig-Zag Region Pulse Structure - Run %03i'%run,fontsize=18)
plt.colorbar(matpic)#,cax=pos)
ax.xaxis.set_ticks_position('bottom')
plt.savefig('zigzag_amp_run%03i_%03i_%03i.pdf'%(run,sstart,send))
plt.show()
```

```python
plt.rcParams.update({'font.size': 18})
fig,ax = plt.subplots(figsize=(10,8))
plt.hist(adc_max,bins=np.linspace(0,100,100),facecolor='blue',log='True',label='Zig-Zag Pulses Found: %ik'%int(len(adc_max)/1000.))
plt.xlim([0,100])
plt.ylim([1,100000])
plt.legend(loc=1)
plt.grid()
plt.xlabel('Max Amplitude in Zig-Zag Pulse')
plt.ylabel('Num. Zig-Zag WFs in Evt.',fontsize=18)
plt.title('Zig-Zag Pulse Amplitude in One Event - Run 148')
plt.show()
```

```python
# fraw Fourier Transforms
#for x in xrange(len(ch_arr['larch'])):
ffts = np.zeros(9600/2+1)
for x in xrange(len(ch_arr['larch'])): 
    starts = ch_arr['zigzag_start'][x]
    ends   = ch_arr['zigzag_end'][x]
    vals = np.zeros(9600)
    for n in xrange(len(starts)):
            for j in xrange(ends[n]-starts[n]):
                vals[starts[n]+j] = 1
    
    ffts  += np.fft.rfft(vals)

ffts = np.abs(ffts)
ffts/=float(len(ch_arr['larch']))
freq = np.fft.rfftfreq(vals.size,0.5)
```

```python
fig,ax = plt.subplots(figsize=(15,10))
timespace = 1/freq
#plt.plot(freq*1000,ffts,'r-')
plt.plot(timespace,ffts,'r-')
plt.ylim([0,100])
plt.xlim([0,100])
plt.xlabel("Period [usec]")
plt.ylabel("FFT Mag.")
plt.show()
```
